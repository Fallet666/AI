4.3.2 Задание
Дана функция sqrt(1 + power(np.e, np.sqrt(x)) + cos(x**2)) / abs(1 - np.sin(x)**3) + log(abs(2*x));
Создать массив из 10 значений функции (𝑥, например, изменяется от 1 до 10). Выделить срез первой половины массива и построить графики для основного массива – линейный и для среза – точечный

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(1, 10, 10)
y = np.sqrt(1 + np.power(np.e, np.sqrt(x)) + np.cos(x**2)) / np.abs(1 - np.sin(x)**3) + np.log(np.abs(2*x))
x_slice = x[:5]  
y_slice = y[:5]
plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Основной массив', color='blue')
plt.scatter(x_slice, y_slice, label='Срез первой половины', color='red')
plt.title('Графики функции y(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

4.3.3 Задание 
Дана функция abs(cos(x*e**(cos(x)+ln(x+1))
Построить график на интервале (0,10) с шагом 1 с заливкой площади и найти эту площадь под ним. Для вычисления площади воспользуйте функции trapz(), вычисляющей интеграл по правилу трапеции. Для ее корректной работы необходимо подключить следующие библиотеки: 
from scipy.integrate import simps 
from numpy import trapz

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.abs(np.cos(x * np.exp(np.cos(x) + np.log(x + 1))))

x = np.arange(0, 10, 1)
y = f(x)

plt.figure(figsize=(10, 5))
plt.plot(x, y, 'b-', label='f(x)')
plt.fill_between(x, y, color='lightblue', alpha=0.3)
plt.title('График функции')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()

area = np.trapz(y, x)
print("Площадь под графиком:", area)

plt.show()
```

4.3.4 Задание
Откройте статистические данные по стоимости акций Apple, Microsoft, Google за 12 месяцев 2021 г. Постройте 3 графика на плоскости и оцените их динамику. Подумайте, как можно улучшить читаемость результатов.

```python
import yfinance as yf
import matplotlib.pyplot as plt

tickers = ['AAPL', 'MSFT', 'GOOGL']

try:
    data = yf.download(tickers, start="2021-01-01", end="2021-12-31")['Close']

    plt.figure(figsize=(12, 6))

    if 'AAPL' in data.columns:
        plt.plot(data.index, data['AAPL'], label='Apple', color='blue')

    if 'MSFT' in data.columns:
        plt.plot(data.index, data['MSFT'], label='Microsoft', color='green')

    if 'GOOGL' in data.columns:
        plt.plot(data.index, data['GOOGL'], label='Google', color='red')

    plt.title('Динамика стоимости акций Apple, Microsoft, Google в 2021 году')
    plt.xlabel('Дата')
    plt.ylabel('Цена закрытия ($)')
    plt.legend()
    plt.grid(True)
    plt.xticks(rotation=45)

    plt.tight_layout()
    plt.show()

except Exception as e:
    print("Произошла ошибка при загрузке данных:", e)
```

4.3.5 Задание
Создайте простейший калькулятор, включающий основные действия
для двух переменных ′ + ′,′ − ′,′ ∙ ′,′ ÷ ′, а также вычисление следующих функций:
e**(x+y), sin(x+y), cos(x+y), x**y

```python
import math

def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y == 0:
        return "Ошибка: деление на ноль"
    return x / y

def exponent(x, y):
    return math.exp(x + y)

def sine(x, y):
    return math.sin(x + y)

def cosine(x, y):
    return math.cos(x + y)

def power(x, y):
    return x ** y

def calculator(x, y, operator):
    operations = {
        '+': add,
        '-': subtract,
        '*': multiply,
        '/': divide,
        'e': exponent,
        'sin': sine,
        'cos': cosine,
        '**': power,
    }
    return operations.get(operator, lambda: "Некорректный оператор")(x, y)

x = float(input("Введите значение x: "))
y = float(input("Введите значение y: "))
operator = input("Введите оператор (+, -, *, /, e, sin, cos, **): ")

result = calculator(x, y, operator)
print("Результат:", result)
# вводил x=10, y=20, оператор e
```
