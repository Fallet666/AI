–¢—Ä–µ–±—É–µ—Ç—Å—è –ø—Ä–æ–≤–µ—Å—Ç–∏ –∫–ª–∞—Å—Ç–µ—Ä–Ω—ã–π –∞–Ω–∞–ª–∏–∑ –¥–∞–Ω–Ω—ã—Ö –º–µ—Ç–æ–¥–æ–º k-—Å—Ä–µ–¥–Ω–∏—Ö.
–ü–æ—ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∏—Ä—É–π—Ç–µ —Å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –∫–ª–∞—Å—Ç–µ—Ä–æ–≤.

```python
import sns
from matplotlib import pyplot as plt
from sklearn.svm._libsvm import predict
from sklearn.cluster import KMeans
import numpy as np

x = np.array([[5,3],
               [10,15],
               [15,12],
               [24,10],
               [30,45],
               [85,70],
               [71,80],
               [60,78],
               [55,52],
               [80,91], ])
kmeans = KMeans (n_clusters = 3)
kmeans. fit(x)
y_kmeans = kmeans.predict(x)
plt.scatter(x[:, 0], x[:, 1], c = y_kmeans, s = 20)
centers = kmeans. cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c = 'blue', s = 100, alpha = 0.9);
plt.show()
```

–í—ã–ø–æ–ª–Ω–∏—Ç–µ –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏—é –¥–ª—è –Ω–∞–±–æ—Ä–∞ –¥–∞–Ω–Ω—ã—Ö –∏—Ä–∏—Å–æ–≤ –§–∏—à–µ—Ä–∞. –í—ã–ø–æ–ª–Ω–∏—Ç–µ
–ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –¥–ª—è –º–æ–¥–µ–ª–∏. –ü–æ—ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∏—Ä—É–π—Ç–µ —Å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –∫–ª–∞—Å—Ç–µ—Ä–æ–≤.

```python
from sklearn.datasets import load_iris
import seaborn as sns
from sklearn.cluster import KMeans
iris = load_iris()
X = iris.data
y = iris.target
n_clusters_list = [2, 3, 4]

# –°–æ–∑–¥–∞–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞
fig, axs = plt.subplots(1, len(n_clusters_list), figsize=(15, 5))

for i, n_clusters in enumerate(n_clusters_list):
    # –°–æ–∑–¥–∞–Ω–∏–µ –∏ –æ–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏ KMeans
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(X)

    # –ü–æ–ª—É—á–µ–Ω–∏–µ –º–µ—Ç–æ–∫ –∫–ª–∞—Å—Ç–µ—Ä–æ–≤
    labels = kmeans.labels_

    # –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
    axs[i].scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
    axs[i].scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='x', c='r', s=100)
    axs[i].set_title(f'{n_clusters} –∫–ª–∞—Å—Ç–µ—Ä–∞')
    axs[i].set_xlabel(iris.feature_names[0])
    axs[i].set_ylabel(iris.feature_names[1])

plt.tight_layout()
plt.show()
```

–ó–∞–¥–∞–Ω–∏–µ. –í—ã–ø–æ–ª–Ω–∏—Ç–µ –∏–µ—Ä–∞—Ä—Ö–∏—á–µ—Å–∫—É—é –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏—é –¥–ª—è –Ω–∞–±–æ—Ä–∞ –¥–∞–Ω–Ω—ã—Ö –æ–±
–∏—Ä–∏—Å–∞—Ö –§–∏—à–µ—Ä–∞. –ü—Ä–∏ —ç—Ç–æ–º –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –ª—é–±—ã–µ –¥–≤–∞ –ø—Ä–∏–∑–Ω–∞–∫–∞
(–≤—Å–µ–≥–æ –∏—Ö —á–µ—Ç—ã—Ä–µ).

```python
from sklearn.datasets import load_iris
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

# –ó–∞–≥—Ä—É–∑–∫–∞ –¥–∞–Ω–Ω—ã—Ö
iris = load_iris()
X = iris.data[:, :2]  # –ò—Å–ø–æ–ª—å–∑—É–µ–º —Ç–æ–ª—å–∫–æ –ø–µ—Ä–≤—ã–µ –¥–≤–∞ –ø—Ä–∏–∑–Ω–∞–∫–∞ (–¥–ª–∏–Ω–∞ –∏ —à–∏—Ä–∏–Ω–∞ —á–∞—à–µ–ª–∏—Å—Ç–∏–∫–∞)

# –ò–µ—Ä–∞—Ä—Ö–∏—á–µ—Å–∫–∞—è –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏—è
agg_clustering = AgglomerativeClustering(n_clusters=3)  # –£–∫–∞–∑—ã–≤–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–ª–∞—Å—Ç–µ—Ä–æ–≤
agg_clustering.fit(X)

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=agg_clustering.labels_, cmap='viridis')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Hierarchical Clustering of Iris Dataset')
plt.colorbar(label='Cluster')
plt.show()
```
