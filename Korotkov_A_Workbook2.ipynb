Создать 8x8 матрицу и заполнить её в шахматном порядке нулями и единицами.

1.3.4 Задание
Создать матрицу с 0 внутри, и 1 на границах.

```python
import numpy as np
rows, cols = 5, 5

matrix = np.zeros((rows, cols), dtype=int)

matrix[0, :] = 1
matrix[-1, :] = 1 
matrix[:, 0] = 1
matrix[:, -1] = 1 

print(matrix)
```

3.3.2 Задание
Загрузить фрейм данных по ссылке:
https://raw.githubusercontent.com/akmand/datasets/master/iris.csv.
Необходимо выполнить нормализацию первого числового признака
(sepal_length_cm) с использованием минимаксного преобразования, а
второго (sepal_width_cm) с задействованием z-масштабирования.

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler

url = "https://raw.githubusercontent.com/akmand/datasets/master/iris.csv"
df = pd.read_csv(url)

print("Первые несколько строк фрейма данных:")
print(df.head())

# Нормализация первого числового признака (sepal_length_cm) с использованием минимаксного преобразования
df['sepal_length_cm'] = MinMaxScaler().fit_transform(df[['sepal_length_cm']])

# Нормализация второго числового признака (sepal_width_cm) с использованием z-масштабирования
df['sepal_width_cm'] = StandardScaler().fit_transform(df[['sepal_width_cm']])

print("\nПервые нескольких строк фрейма данных после нормализации:")
print(df.head())
```
